```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Baca file CSV
data = pd.read_csv('C:/Users/Administrator/Documents/FITRI/PEMOGRAMAN/H05.csv')
```

```python
# 1. Histogram nilai matematika
plt.figure(figsize=(8, 5))
plt.hist(data['nilai_mat'], bins=8, color="pink", edgecolor="black")
plt.title('Histogram Nilai Matematika')
plt.xlabel('Nilai')
plt.ylabel('Jumlah Mahasiswa')
plt.show()
```

```python
# 2. Diagram batang horizontal jumlah mahasiswa per fakultas
plt.figure(figsize=(8, 5))
data['fakultas'].value_counts().plot(kind='barh', color="red", edgecolor="black")
plt.title('Jumlah Mahasiswa per Fakultas')
plt.xlabel('Jumlah Mahasiswa')
plt.ylabel('Fakultas')
plt.show()
```

```python
# 3. Diagram pie jumlah mahasiswa per fakultas
plt.figure(figsize=(8, 8))
data['fakultas'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Presentase Mahasiswa per Fakultas')
plt.show()
```

```python
# 4. Berdasarkan diagram pie dan batang horizontal manakah fakultas dengan mahasiswa terbanyak?
# Manakan diagram yang lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak dan mengapa?
# Jawaban: Fakultas dengan mahasiswa terbanyak adalah Fakultas A.
# Diagram batang horizontal lebih baik karena lebih mudah membandingkan jumlah mahasiswa dari masing-masing fakultas.
```

```python
# 5. Scatter plot nilai kimia dan fisika
plt.figure(figsize=(8, 5))
plt.scatter(data['nilai_kim'], data['nilai_fis'])
plt.title('Hubungan Nilai Kimia dan Fisika')
plt.xlabel('Nilai Kimia')
plt.ylabel('Nilai Fisika')
plt.show()
```
