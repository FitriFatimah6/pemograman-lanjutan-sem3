```python
file_path = 'C:/Users/Administrator/Documents/PEMOGRAMAN/H05.csv'

try: 
    df = pd.read_csv(file_path)
    
    # 1. Hitung total data
    print("\n1. Total jumlah siswa:", len(df))
    
    # 2. Tampilkan 10 data pertama
    print("\n2. 10 siswa pertama:")
    print(df.head(10))
    
    # 3. Tampilkan data dari indeks 50-60
    print("\n3. Siswa dari indeks 50 hingga 60:")
    print(df.iloc[49:60])  # Menggunakan 49 karena indeks mulai dari 0
    
    # 4. Hitung siswa per fakultas
    print("\n4. Siswa per fakultas:")
    faculty_counts = df['fakultas'].value_counts()
    print(faculty_counts)
    
    # 5. Hitung korelasi antara fisika dan kimia
    korelasi = df['nilai_fis'].corr(df['nilai_kim'])
    print("\n5. Korelasi antara Fisika dan Kimia:", korelasi)
    
    # Menentukan jenis korelasi
    print("Nilai korelasi antara Fisika dan Kimia:", korelasi)
    if korelasi > 0.7:
        print("Korelasi kuat dan lurus.")
    elif korelasi < -0.7:
        print("Korelasi kuat dan terbalik.")
    elif -0.7 <= korelasi <= 0.7:
        print("Tidak ada korelasi yang signifikan.")

except FileNotFoundError:
    print(f"Kesalahan: File di {file_path} tidak ditemukan.")
except pd.errors.EmptyDataError:
    print("Kesalahan: File kosong.")
except pd.errors.ParserError:
    print("Kesalahan: Ada masalah saat mem-parsing file.")
except KeyError as e:
    print(f"Kesalahan: Kolom {e} tidak ada dalam DataFrame.")
except Exception as e:
    print(f"Terjadi kesalahan yang tidak terduga: {e}")
```

```python
import pandas as pd

file_path = 'C:/Users/Administrator/Documents/PEMOGRAMAN/H05.csv'
try:
    df = pd.read_csv(file_path)
    
    # 1. Tampilkan nilai matematika, fisika, dan kimia milik Tuan Leo
    print("\n1. Nilai Tuan Leo:")
    leo_data = df[df['nama'] == 'Leonardus Kevin'][['nilai_mat', 'nilai_fis', 'nilai_kim']]
    if not leo_data.empty:
        print(leo_data)
    else:
        print("Data Leonardus Kevin tidak ditemukan")
    
    # 2. Cari mahasiswa dengan nilai fisika tertinggi
    print("\n2. Mahasiswa dengan nilai fisika tertinggi:")
    max_fisika = df[df['nilai_fis'] == df['nilai_fis'].max()]
    print(max_fisika[['nama', 'nilai_fis']])
    
    # 3. Tampilkan 10 mahasiswa dengan nilai kimia tertinggi
    print("\n3. 10 mahasiswa dengan nilai kimia tertinggi:")
    top_10_kimia = df.nlargest(10, 'nilai_kim')[['nama', 'nilai_kim']]
    print(top_10_kimia)
    
    # 4. Hitung banyaknya nilai matematika di bawah 50
    nilai_mat_dibawah_50 = len(df[df['nilai_mat'] < 50])
    print("\n4. Jumlah mahasiswa dengan nilai matematika di bawah 50:", nilai_mat_dibawah_50)
    
    # 5. Hitung banyaknya mahasiswa di FMIPA
    jumlah_mhs_fmipa = len(df[df['fakultas'] == 'FMIPA'])
    print("\n5. Jumlah mahasiswa di FMIPA:", jumlah_mhs_fmipa)

except FileNotFoundError:
    print(f"Kesalahan: File di {file_path} tidak ditemukan.")
except pd.errors.EmptyDataError:
    print("Kesalahan: File kosong.")
except pd.errors.ParserError:
    print("Kesalahan: Ada masalah saat mem-parsing file.")
except KeyError as e:
    print(f"Kesalahan: Kolom {e} tidak ada dalam DataFrame.")
except Exception as e:
    print(f"Terjadi kesalahan yang tidak terduga: {e}")
```
